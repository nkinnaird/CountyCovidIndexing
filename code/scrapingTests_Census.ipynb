```python
from bs4 import BeautifulSoup
import requests
```

```python
url_census = 'https://www.census.gov/quickfacts/fact/table/adamscountyillinois/'

census_response = requests.get(url_census)
print(census_response.status_code)
census_page = census_response.text
```

```python
census_soup = BeautifulSoup(census_page, "lxml")
# print(census_soup)
census_table = census_soup.find(id='detailedData')
print(census_table)
```
