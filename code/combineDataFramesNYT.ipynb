```python
# load census data and drop the District of Columbia
county_census_info = pd.read_pickle("../Data/county_census_info.pkl")
# county_census_info.drop('District of Columbia', level='STATE', inplace=True)
county_census_info.info()
```

```python
county_census_info.set_index('FIPS Code', inplace=True)
county_census_info.head(5)
```

```python
covid_df = pd.read_csv('../Data//NYT-Covid/us-counties-totals.csv')
covid_reduced_df = covid_df[['county', 'state', 'fips', 'cases', 'deaths']]
covid_reduced_df.set_index('fips', inplace=True)
covid_reduced_df
```

```python
# join the census and covid dataframes
full_df = pd.merge(county_census_info, covid_reduced_df, left_index=True, right_index=True)
```
