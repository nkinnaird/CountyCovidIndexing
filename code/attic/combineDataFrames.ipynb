```python
import math
import pandas as pd
import glob
```

```python
# load census data and drop the District of Columbia
county_census_info = pd.read_pickle("../Data/county_census_info.pkl")
county_census_info.drop('District of Columbia', level='STATE', inplace=True)
county_census_info.info()
```

```python
# check state data frames
state_pkls = glob.glob("../Data/Covid/*.pkl")
sorted(state_pkls)

size_check = 0
for state in state_pkls:
    df_temp = pd.read_pickle(state)
    size_check += len(df_temp.index)
    print(state, len(df_temp.index))
print(size_check)
```

```python
# combine state data frames
county_covid_info = pd.concat([pd.read_pickle(state) for state in sorted(state_pkls)], axis=0, join='outer')
county_covid_info.reset_index()
county_covid_info.info()
```

```python
county_covid_info
```

```python
county_census_info
```

```python
# join the census and covid dataframes
full_df = county_census_info.join(county_covid_info, how='outer')
```

```python
full_df.to_pickle("../Data/county_data_all.pkl")
```

```python
# drop columns which are not of interest - I can come back here and change my cuts if I think they might be useful later on
full_df.columns
# column_indexes_to_keep = [0, 6, 18, 20, 25, 26, 31, 32, 33, 34, 38, 43, 44, 46, 48, 59, 62, 63, 64, 65, 66, 67, 68 , 69, 70, 71, 72]
column_indexes_to_keep = [0, 6, 18, 20, 25, 26, 31, 32, 33, 34, 38, 43, 44, 46, 48, 59, 64, 65]
columns_to_keep = []
for col in full_df.columns:
    col_index = full_df.columns.get_loc(col)
    if(col_index in column_indexes_to_keep): columns_to_keep.append(col)
#     print(full_df.columns.get_loc(col), col)
    
# columns_to_keep
reduced_df = full_df[columns_to_keep]
reduced_df.info()
```
